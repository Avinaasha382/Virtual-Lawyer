```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json('Final.json')
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Number of Citations'])

plt.title('Box Plot of Citations')
plt.xlabel('Amount')

plt.show()
```

```python
# Calculating Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Number of Citations'].quantile(0.25)
Q3 = df['Number of Citations'].quantile(0.75)

# Calculating IQR
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers: Values outside the bounds
outliers = df[(df['Number of Citations'] < lower_bound) | (df['Number of Citations'] > upper_bound)]

print(lower_bound)
print(upper_bound)
# Display outliers
print("Outliers detected:")
print(outliers['case_id'])
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier  # You can choose any classifier
from sklearn.metrics import accuracy_score, classification_report
columns_to_extract = ['Paragraph Count', 'Word Count', 'Average Sentence Length','Bench Composition', 'Relief Granted', 'Number of Citations','Number of Primary Statutes', 'Number of Core Legal Issues', 'Case Type']

# Extract these columns from the DataFrame
extracted_df = df[columns_to_extract]

# Display the resulting DataFrame
#print(extracted_df)
# Step 1: Filter the DataFrame to include only rows with classes 'c1' or 'c2'
filtered_df = extracted_df[df['Case Type'].isin(['Civil Appeal', 'Criminal'])]

# Step 2: Separate features and labels
X = filtered_df.drop(columns=['Case Type'])  # 8 columns as features
y = filtered_df['Case Type']                 # Target variable (class labels)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train a classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Step 5: Make predictions on the test set and evaluate
y_pred = clf.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
